```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("crime_dataset_india.csv")
df.drop_duplicates(inplace=True)
df.head()
```

```python
df["Date of Occurrence"]=pd.to_datetime(df["Date of Occurrence"],errors="coerce",format="mixed")
df["Date Reported"]=pd.to_datetime(df["Date Reported"],errors="coerce",format="mixed")
df["Time of Occurrence"]=pd.to_datetime(df["Time of Occurrence"],errors="coerce",format="mixed")
df["Date Case Closed"]=pd.to_datetime(df["Date Case Closed"],errors="coerce",format="mixed")
```

```python
df["Year"]=df["Date of Occurrence"].dt.year
df["Month"]=df["Date of Occurrence"].dt.month
df["Hour"]=df["Time of Occurrence"].dt.hour
df["Case Duration"]=(df["Date Case Closed"]-df["Date Reported"]).dt.days
```

```python
df["Weapon Used"]=df["Weapon Used"].fillna("Unknown")
df["Case Status"]=df["Date Case Closed"].apply(lambda x:"Open" if pd.isna(x) else "Closed")
```

```python
def age_group(age):
  if age<18:
    return "Teen"
  elif age<35:
    return "Young Adult"
  elif age<60:
    return "Adult"
  else:
    return "Senior"

df["Age Group"]=df["Victim Age"].apply(age_group)
```

```python
def time_peroid(hour):
  if hour<6:
    return "Late Night"
  elif hour<12:
    return "Morning"
  elif hour<18:
    return "Afternoon"
  else:
    return "Night"

df["time_peroid"]=df["Hour"].apply(time_peroid)
```

```python
df["Crime Description"].unique()
```

```python
def severity(crime):
  high=[
      'HOMICIDE','SEXUAL ASSAULT','KIDNAPPING','DOMESTIC VIOLENCE',
      'FIREARM OFFENSE'
  ]
  medium=[
     'ASSAULT','ROBBERY','BURGLARY','ARSON','EXTORTION','VEHICLE - STOLEN',
     'DRUG OFFENSE'
  ]
  if crime in high:
    return "High"
  elif crime in medium:
    return "Medium"
  else:
    return "Low"

df["Crime Severity"]=df["Crime Description"].apply(severity)
```

```python
def closure_speed(days):
  if pd.isna(days):
    return "Open"
  elif days<=30:
    return "Fast"
  elif days<=90:
    return "Average"
  else:
    return "Slow"
df["Closure Speed"]=df["Case Duration"].apply(closure_speed)
```

```python
city_score=df.groupby("City")["Report Number"].count().reset_index()
city_score.columns=["city","crime_count"]
city_score.head()
```

```python
city_score["risk_score"]=(
    city_score["crime_count"]/city_score["crime_count"].max()
)*100
```

```python
def risk_category(score):
  if score>=75:
    return "High Risk"
  elif score>=40:
    return "Medium Risk"
  else:
    return "Low Risk"
city_score["risk_category"]=city_score["risk_score"].apply(risk_category)
```

```python
city_score.head()
```

```python
df=df.merge(
    city_score[["city","risk_score","risk_category"]],
    left_on='City',
    right_on='city',
    how='left'
)
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```
